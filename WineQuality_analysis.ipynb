```python
wq_qualiti=wq[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol', 'quality']].groupby('quality').sum()
wq_qualiti.head()
```

```python
wq_qualiti.reset_index(level=0,inplace=True)
wq_qualiti.head()
```

```python
wq
```

```python
plt.figure(figsize=(25,8))
plt.plot(wq_qualiti['quality'],wq_qualiti['free sulfur dioxide'])
plt.title('Effect of Free sulfur Dioxide on Quality of Wine ',fontsize=25)
plt.xlabel('Quality',fontsize=25)
plt.ylabel('Free sulfur Dioxide',fontsize=25)
```

```python
plt.figure(figsize=(25,8))
plt.plot(wq_qualiti['quality'],wq_qualiti['total sulfur dioxide'])
plt.title('Effect of total sulfur Dioxide on Quality of Wine ',fontsize=25)
plt.xlabel('Quality',fontsize=25)
plt.ylabel('Total sulfur Dioxide',fontsize=25)
```

```python
 wq_quality= wq[['quality','total sulfur dioxide','free sulfur dioxide']].groupby('quality').sum()
wq_quality.reset_index(level=0,inplace=True)
wq_quality.sort_values(by='total sulfur dioxide',ascending=False,inplace=True)
wq_quality.head()
```

```python
plt.figure(figsize=(20,7))
sns.barplot(x=wq_quality['quality'],y = wq_quality['total sulfur dioxide'])
plt.title('Effect of Total Sulfur Dioxide on Quality',fontsize =20)
plt.xlabel('quality',fontsize=20)
plt.ylabel('Total Sulfur Dioxide',fontsize=20)
```

```python
wq_quality_max = wq[['quality','total sulfur dioxide']].groupby('quality').max()
wq_quality_max.reset_index(level=0,inplace=True)
wq_quality_max.columns = ['Quality','Max SO2']
wq_quality_max.head()
```

```python
wq_quality_min = wq[['quality','total sulfur dioxide']].groupby('quality').min()
wq_quality_min.reset_index(level=0,inplace=True)
wq_quality_min.columns = ['Quality','Min SO2']
wq_quality_min.head()
```

```python
st_quality = pd.merge(wq_quality_max,wq_quality_min,how='inner',on='Quality')
st_quality.head()
```

```python
st_quality['Per_Change'] = ((st_quality['Max SO2']-st_quality['Min SO2'])/st_quality['Max SO2'])*100
st_quality.sort_values(by='Per_Change',ascending=False,inplace=True)
st_quality.head()
```

```python
plt.figure(figsize=(20,7))
sns.barplot(x='Quality',y='Per_Change',data= st_quality)
plt.title('SO2 Effect Trend',fontsize=20)
plt.xlabel("Quality",fontsize=15)
plt.ylabel("% Decline",fontsize=15)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(wq.drop(['total sulfur dioxide', 'quality'],axis=1), 
                                                    wq['total sulfur dioxide'], test_size=0.20, 
                                                    random_state=8)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
accuracy = model.score(X_test, y_test)
print(accuracy*100,'%')
```
